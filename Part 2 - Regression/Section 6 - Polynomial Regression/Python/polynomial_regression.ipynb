# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values    # Only take Level column
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset
(Not enough data to split into training and test set)

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
print(X_poly)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results
This give a poor fit to the data. For levels 4-8 the offered salary would be too high and levels 1, 2 and 10 too low.

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Salary truth or bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Salary truth or bluff (Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution)

```
poly_reg = PolynomialFeatures(degree=4)     # higher degree polynomial
X_poly_hi = poly_reg.fit_transform(X)
print(X_poly_hi)

lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_hi, y)
```

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_3.predict(X_poly_hi), color='blue')
plt.title('Salary truth or bluff (Higher Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```
y_pred = lin_reg.predict([[6.5]])
print(y_pred)

# Predicted salary is too high!
```

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.scatter(6.5, y_pred, color='green')

plt.title('Salary truth or bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Polynomial Regression

```
y_pred_poly = lin_reg_3.predict(poly_reg.fit_transform([[6.5]]))    # 1x1 matrix
print(y_pred_poly)
```

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_3.predict(X_poly_hi), color='blue')
plt.scatter(6.5, y_pred_poly, color='green')

plt.title('Salary truth or bluff (Higher Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
